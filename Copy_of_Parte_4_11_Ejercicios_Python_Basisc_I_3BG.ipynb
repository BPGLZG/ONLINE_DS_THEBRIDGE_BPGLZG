## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
print(A)
B = "Text"
print(B)
C = 4.1
print(C)

# 1. Si A y B son equivalentes
print(A == B)

# 2. Si A y C no son equivalentes
print(A != C)

# 3. Si A es mayor que C
print(A > C)

# 4. Si C es menor o igual que A
print(C <= A)

# 5. Si B no es equivalente a C
print(B != C)
````

## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
# Pedir datos al usuario
valor1 = input("Introduce el primer valor: ")
valor2 = input("Introduce el segundo valor: ")

# Comparar y mostrar resultado
if valor1 == valor2:
    print("Los valores son iguales.")
else:
    print("Los valores son diferentes.")
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
# Pedir datos al usuario
valor1 = input("Introduce el primer valor: ")
valor2 = input("Introduce el segundo valor: ")
valor3 = input("Introduce el tercer valor: ")

# Comprobar si todos son iguales
if valor1 == valor2 == valor3:
    print("Todos los valores son iguales.")
else:
    print("No todos los valores son iguales.")

# Comprobar si al menos dos son iguales
if valor1 == valor2 or valor1 == valor3 or valor2 == valor3:
    print("Al menos dos valores son iguales.")
else:
    print("Ningún par de valores es igual.")
````
