`````python
def buscar_libro(libreria, titulo, idioma="ninguno"):
    for libro in libreria:
        if libro["titulo"].lower() == titulo.lower():
            if idioma == "ninguno" or libro["idioma original"].lower() == idioma.lower():
                return libro
    return "Ese no lo tengo, ¿mola?"
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def dni_valido(dni):
    palabra = "TRWAGMYFPDXBNJZSQVHLCKE"

    # Comprobar formato básico: 7 u 8 números + 1 letra
    if len(dni) < 8 or len(dni) > 9:
        return False

    numeros = dni[:-1]
    letra = dni[-1].upper()

    if not numeros.isdigit():
        return False

    dni_num = int(numeros)
    letra_correcta = palabra[dni_num % 23]

    return letra == letra_correcta
`````
