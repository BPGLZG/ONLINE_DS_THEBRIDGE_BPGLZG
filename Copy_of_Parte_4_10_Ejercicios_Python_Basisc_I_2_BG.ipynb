## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
# Definir pi manualmente
pi = 3.1416

# Pedir al usuario los grados
grados = float(input("Introduce el valor en grados: "))

# Convertir a radianes
radianes = grados * (pi / 180)

# Mostrar el resultado
print("El valor en radianes es:", radianes)
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
# Pedir datos al usuario
base = float(input("Introduce la base: "))
altura = float(input("Introduce la altura: "))

# Calcular el área
area = base * altura

# Mostrar el resultado
print("El área del paralelogramo es:", area)
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
# Pedir datos al usuario (pueden usar coma como decimal)
base = input("Introduce la base: ")
altura = input("Introduce la altura: ")

# Convertir a número (cambiando coma por punto)
base = float(base.replace(",", "."))
altura = float(altura.replace(",", "."))

# Calcular el área
area = base * altura

# Mostrar resultado
print("El área del paralelogramo es:", area)
```
